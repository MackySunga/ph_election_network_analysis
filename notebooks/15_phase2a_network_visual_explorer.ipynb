## How to read the graphs

| Visual element | Meaning |
|---|---|
| Node | Candidate, hashtag, or network actor |
| Edge / line | Relationship between nodes |
| Thicker edge | Stronger relationship / higher co-occurrence |
| Larger node | Higher network importance or centrality |
| Color | Community, node type, or online/election alignment category |
| Dark node border | Actual Top 12 Senate winner in the overlay graph |

The most important interpretation is that this project does not only count tweets. It models election discourse as a **connected political communication system**.
